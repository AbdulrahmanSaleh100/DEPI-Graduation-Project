```python
import pandas as pd

# Load datasets
df_symptoms = pd.read_csv("Disease and symptoms dataset.csv")   # الملف الكبير
df_medicine = pd.read_csv("merged_all.csv")   # ملف الأدوية

# تنظيف أسماء الأمراض لضمان التطابق الكامل
df_symptoms['diseases'] = (
    df_symptoms['diseases']
    .astype(str)
    .str.strip()
    .str.lower()
)

df_medicine['diseases'] = (
    df_medicine['diseases']
    .astype(str)
    .str.strip()
    .str.lower()
)

# Merge جميع الأعمدة من الملفين
merged_df = pd.merge(
    df_symptoms,
    df_medicine,
    on='diseases',
    how='left'
)

# الأعمدة المطلوب وضعها بعد Category
medicine_columns = [
    'medicine_1',
    'medicine_2',
    'medicine_3',
    'side_effects'
]

# ترتيب الأعمدة
cols = merged_df.columns.tolist()

# إزالة الأعمدة مؤقتًا
for col in medicine_columns:
    if col in cols:
        cols.remove(col)

# تحديد مكان Category
category_index = cols.index('Category') + 1

# إدراج الأعمدة بعد Category
new_cols = (
    cols[:category_index]
    + medicine_columns
    + cols[category_index:]
)

# إعادة ترتيب الداتا
merged_df = merged_df[new_cols]

# حفظ الملف النهائي
merged_df.to_csv("final_merged_dataset_1.csv", index=False)

print("Merge completed successfully ✅")
print("Final Shape:", merged_df.shape)

# عرض أول 5 صفوف
print(merged_df.head())
```
